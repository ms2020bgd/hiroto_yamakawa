```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import time#

import pandas as pd
```

```python
url = "https://gist.github.com/paulmillr/2657075"
```

```python
#Récupérer via crawling la liste des 256 top contributors sur cette page https://gist.github.com/paulmillr/2657075

def github_crawling(url):
 
    page = requests.get(url, verify=True, timeout = 5)
    html = page.text
    soup = BeautifulSoup(html, "html.parser")

    divs = soup.find('table').find('tbody').find_all('tr')
    List_users=[]
    for index, line in enumerate(divs):
        name = line.find('td').text.split(" ")[0]
        List_users.append(name)
    
    return List_users
```

```python
github_crawling(url)
```

```python
def get_mean_star_repo(user):
    
    url = f'https://api.github.com/users/{user}/repos'
    headers = {'Authorization': 'token 9cc4c16860b793069dc315fcfd5bf8585ade0f16'}
    #to authenticate with a token
    res = requests.get(url, headers=headers)
    repositories = res.json()
    
    #sum stars for all the repos and divide by the number of repo : len(repository)
    total_star = 0
    for repository in repositories:
        total_star += repository['stargazers_count']
        
    #to avoid null case 
    if len(repository) == 0:
        return 0
    
    mean_star = total_star / len(repository)

    return mean_star
```

```python
def top_users(url):
    list = []
    for user in github_crawling(url)[0:10]: #limited to 10 for that example
        info_user = {'user': user, 'mean_star' : get_mean_star_repo(user) }
        list.append(info_user)
    return list
```

```python
users = pd.DataFrame(top_users(url))
```
