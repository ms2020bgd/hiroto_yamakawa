```python
import requests
import urllib.request
from bs4 import BeautifulSoup
import time
```

```python
#Test URL and TARGET
url = "https://en.wikipedia.org/wiki/Duquesne_University"
```

```python
def way_to_philosophy(url):
    target = "http://en.wikipedia.org/wiki/Philosophy"
    count = 0

    while url != target :
        page = requests.get(url, verify=True, timeout = 5)
        html = page.text
        soup = BeautifulSoup(html, "html.parser")

        divs = soup.find(id="mw-content-text").find('div',{'class': "mw-parser-output"}).find_all('p')

        Links=[]
    
        for div in divs:                         #iterate over the list of <p>
            for link in div.find_all(href= True):   #a link is /wiki/something
                Links.append(link['href'])
                
        next = Links[0]    #
        url ="http://en.wikipedia.org" + next
        count +=1
        print(url)
        time.sleep(1)

    print(count)
                
```

```python
way_to_philosophy(url)
```

```python
def way_to_philosophy_2(url):
    
    target = "http://en.wikipedia.org/wiki/Philosophy"
    count = 0
    history = []      #add a list "History" to avoid going back to a link previously visited 
    while url != target :
        page = requests.get(url, verify=True, timeout = 5)
        html = page.text
        soup = BeautifulSoup(html, "html.parser")

        divs = soup.find(id="mw-content-text").find('div',{'class': "mw-parser-output"}).find_all('p')

        Links=[]
    
        for div in divs:
            for link in div.find_all(href= True):
                Links.append(link['href'])
                
        filtered_links = []
        
        for i in Links :      # Note for improvement : filter list rather than conditionnal
            
            if i in history:
                continue
            
            elif "Help" in i:
                continue
            
            elif "#cite_note" in i:
                continue
            
            elif "wiktionary" in i:
                continue
            
            elif ":" in i:
                continue
                
            elif "." in i:
                continue
            
            elif "_" in i:    #added to avoid the issue with http://en.wikipedia.org/wiki/Factor_(disambiguation)
                continue   
            
            filtered_links.append(i)
        
        next = filtered_links[0]
        
        history.append(next)
        url ="http://en.wikipedia.org" + next 
        count +=1
        
        print(url)
        time.sleep(0.5)

    print(count)
            
```
